```python
from matplotlib import font_manager
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Image

font_path = 'Inter.ttf' 
font_manager.fontManager.addfont(font_path)
prop = font_manager.FontProperties(fname=font_path)
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = prop.get_name()

plt.rcParams.update({
    "figure.facecolor":  (1.0, 1.0, 1.0, 0.0),  # red   with alpha = 30%
    "axes.facecolor":    (0.0, 0.0, 0.0, 0.1),  # green with alpha = 50%
    "savefig.facecolor": (1.0, 1.0, 1.0, 0.0),  # blue  with alpha = 20%
})

# ---


labels = ['Android', 'Windows', 'iOS', 'MacOS', 'Linux', 'Autre']
values = [43.88, 27.71, 17.13, 6.27, 1.14, 3.87]
colors = ['#66BB6A', '#80D8FF', '#455A64', '#263238', '#FF8A65', '#FFEE58']


fig, ax = plt.subplots(figsize=(3.2, 2.2))
patches, texts, autotexts = ax.pie(values, labels=labels, colors=colors, autopct='%.0f%%',
                                  )

[patch.set_alpha(0.75) for patch in patches]
[text.set_color('white') for text in texts]
# [text.set_horizontalalignment('center') for text in texts]



[autotext.set_color('white') for autotext in autotexts]

x, y = autotexts[-2].get_position()
autotexts[-2].set_position((x + 0.2, y - 0.05))
# ---
# plt.title("Parts de marché des systèmes d'exploitation (Février 2023)", c='white')
FILENAME = "../market.png"
plt.tight_layout()
plt.savefig(FILENAME, dpi=300)
display(Image(filename=FILENAME))
```
